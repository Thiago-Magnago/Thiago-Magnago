Notebook para obtenção dos municípios brasileiros junto a API do IBGE

```python
import json
import requests

#url = 'https://servicodados.ibge.gov.br/api/v1/localidades/regioes/1|2|3|4|5/municipios'
url = 'https://servicodados.ibge.gov.br/api/v1/localidades/municipios'

res = requests.get(url)

#Obtendo o JSON da URL em formato texto (string) 
texto = res.text
print(type(texto))
```

Conversão de dados de string em formato JSON

```python
#O módulo json vem com um método chamado loads(), o s em loads() significa string. 
#Como queremos converter dados de string em JSON, usaremos este método.
obj = json.loads(texto)
print(type(obj))
```

```python
#Quantidade de municípios
print("Há {} municípios no total.".format(len(obj)))
```

Vemos que o obj é do tipo <class 'list'> porque, neste caso, a resposta JSON, como visto, começa com colchetes [ ].

Vamos então normalizar os dados com o json_normalize(), de modo a 'abrir' os registros aninhados.

```python
import pandas as pd

#Normalizando os dados
df = pd.json_normalize(obj)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#Selecionando os campos
municipios_ibge = df[['id', 'nome', 'microrregiao.mesorregiao.UF.nome', 'microrregiao.mesorregiao.UF.sigla', 'regiao-imediata.regiao-intermediaria.UF.regiao.nome', 'regiao-imediata.regiao-intermediaria.UF.regiao.sigla']]
municipios_ibge.head()
```

```python
#Renomeando as colunas

renomear = {
    'nome': 'Municipio',
    'microrregiao.mesorregiao.UF.nome': 'Estado',
    'microrregiao.mesorregiao.UF.sigla': 'UF',
    'regiao-imediata.regiao-intermediaria.UF.regiao.nome': 'Regiao',
    'regiao-imediata.regiao-intermediaria.UF.regiao.sigla': 'Regiao_Sigla'
}

municipios_ibge = municipios_ibge.rename(columns = renomear)
municipios_ibge.head()
```

```python
#Exportando csv

municipios_ibge.to_csv('municipios_ibge.csv', index=False, encoding='utf-8')
```
